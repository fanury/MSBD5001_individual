```python
import pandas as pd
import numpy as np
import time

from sklearn.datasets import make_classification
from sklearn.linear_model import SGDClassifier
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
# fix random seed for reproducibility
seed = 7
np.random.seed(seed)
```

```python
TrainData=pd.read_csv('train.csv')#training data
TestData=pd.read_csv('test.csv')#testing data
heads=[row for row in TrainData]
label=TrainData['time'].values
trainlen=400
testlen=100
```

```python
#获取测试数据对应的时间
realtime1=[]
for i in range(0, trainlen):
    print('%d/'%i+'%d'%trainlen)
    X1,Y1=make_classification(n_samples = TrainData.iat[i,7], n_features= TrainData.iat[i,8] , n_classes = TrainData.iat[i,9]  , n_clusters_per_class = TrainData.iat[i,10] ,n_informative = TrainData.iat[i,11], flip_y= TrainData.iat[i,12] , scale= TrainData.iat[i,13])
    import time
    from sklearn.linear_model import SGDClassifier
    clf = SGDClassifier(penalty = TrainData.iat[i,1], l1_ratio = TrainData.iat[i,2], alpha = TrainData.iat[i,3], max_iter = TrainData.iat[i,4], random_state = TrainData.iat[i,5], n_jobs = TrainData.iat[i,6])
    start = time.time()  
    clf.fit(X1, Y1)
    end=time.time()
    elapsed = (end - start)
    realtime1.append(elapsed)   
```

```python
#获取测试数据的实际时间
testrealtime1=[]
for i in range(0, testlen):
    print('%d/'%i+'%d'%testlen)
    X1,Y1=make_classification(n_samples = TestData.iat[i,7], n_features=TestData.iat[i,8] , n_classes =TestData.iat[i,9]  , n_clusters_per_class = TestData.iat[i,10] ,n_informative = TestData.iat[i,11], flip_y= TestData.iat[i,12] , scale=TestData.iat[i,13])
    import time
    from sklearn.linear_model import SGDClassifier
    clf = SGDClassifier(penalty = TestData.iat[i,1], l1_ratio =TestData.iat[i,2], alpha =TestData.iat[i,3], max_iter =TestData.iat[i,4], random_state =TestData.iat[i,5], n_jobs =TestData.iat[i,6])
    start = time.time()  
    clf.fit(X1, Y1)
    end=time.time()
    elapsed = (end - start)
    testrealtime1.append(elapsed)  
testrealtime1
```

```python
train1=TrainData[heads[1:14]]
train1=pd.get_dummies(train1,columns=[heads[1]])
test1=TestData[heads[1:14]]
test1=pd.get_dummies(test1,columns=[heads[1]])
train1['n_jobs']=train1['n_jobs'].replace(-1,16)
train1['n_jobs']=train1['n_jobs'].fillna(0)
test1['n_jobs']=test1['n_jobs'].replace(-1,16)
test1['n_jobs']=test1['n_jobs'].fillna(0)
```

```python
#label
label=TrainData[heads[-1]]
```

```python
#建立全连接的神经网络
model = Sequential()
model.add(Dense(output_dim=750, input_dim=17, init='uniform', activation='relu'))
model.add(Dropout(0.2))
#model.add(Dense(output_dim=1000, input_dim=1000, init='uniform', activation='relu'))
#model.add(Dropout(0.2))
model.add(Dense(output_dim=1, input_dim=750,init='uniform'))

# Compile model
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
#对数据进行标准化
train1=train1.apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x))) 
test1=test1.apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x))) 

#这里如果能同一度量进行标准化效果可能会更好
#因为实际差距train和test的差距并不是很大，所以分开标准化不会有太大影响
```

```python
#加入实际时间作为feature
realtime=pd.DataFrame(realtime1,columns=['realtime'])
train1['realtime']=realtime
testrealtime=pd.DataFrame(testrealtime1,columns=['realtime'])
test1['realtime']=testrealtime
```

```python
#拟合模型并迭代500次
model.fit(train1, label, nb_epoch=500, batch_size=5,  verbose=2)
# evaluate the model
```

```python
#对test进行预测
y1=model.predict(test1)

result=[]
for i in y1:
    result.append(i[0])

sum=0
n=0
y1=result
for i in range(0,100):
    if y1[i]<0:
        print(i,end=' ')
        sum=sum+label[i]
        n=n+1
        print(y1[i],end=' ')
        print(label[i])
        y1[i]=testrealtime['realtime'][i]
#如果存在负值就将负值转换成label中的值

#将文件写入到csv中并保存
a=pd.DataFrame({'id':TestData['id'],'time':y1})
a.to_csv('prediction1121.csv',columns=['id','time'],index=None)
```
